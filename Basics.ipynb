## Data Types

### Numbers

```python
#integers
a = 10
b = 20  
c = a + b
print("Sum of a and b is:", c)

print(type(a))
```

```python
#floats
x = 5.5
y = 4.5
z = x * y
print("Product of x and y is:", z)

print(type(x))
```

```python
#complex numbers
p = 2 + 3j
q = 1 + 4j
r = p + q
print("Sum of p and q is:", r)

print(type(p))

```

### Booleans


```python
#True or False
is_sunny = True
is_raining = False
print("Is it sunny today?", is_sunny)

print("4<6=" ,4<6)
print("8>10=", 8>10)
print(type(is_sunny))
```

### Strings

```python
#strings
a = "Hello"
b = "World"
c = a + " " + b
print("Concatenated string:", c)
print(type(a))

#using indexing to access characters in a string
print(a[0]) 
print(b[2])

#using slicing to access a substring
print(c[0:5])
print(b[2:])

#Converting cases
print(c.upper())
print(b.lower())

print(c.replace("World", "Python"))
print(c)

b = b.replace("World", "Python")
print(b)

print(a + " " + b)
```

### Type Conversion


```python
#implicit
a = 10
b = 20.2
c = a + b
print("Sum of a and b is:", c)
print(type(c))

#explicit

d = "10"
r = 5

e = int(d)
f = float(d)
g = bool(d)

print(e, type(e))
print(f, type(f))
print(g, type(g))   

print(r + e)
```

## Operators

```python
#arithmatic operators
num1 = 15
num2 = 4
print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)
print("Division:", num1 / num2)
print("Floor Division:", num1 // num2)
print("Modulus:", num1 % num2)  
print( num1 ^ num2)
```
