## Data Types

### Numbers

```python
#integers
a = 10
b = 20  
c = a + b
print("Sum of a and b is:", c)

print(type(a))
```

```python
#floats
x = 5.5
y = 4.5
z = x * y
print("Product of x and y is:", z)

print(type(x))
```

```python
#complex numbers
p = 2 + 3j
q = 1 + 4j
r = p + q
print("Sum of p and q is:", r)

print(type(p))

```
