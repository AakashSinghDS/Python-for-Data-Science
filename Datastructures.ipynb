## Data Structure and Control Flow


### Lists

```python
# Lists (Array in other languages)

# A list is a collection which is ordered and changeable. Allows duplicate members.

list1 = [23, 46, 78, "Hello", "World"]
print(type(list1))
print(list1[3])  # Accessing an element by index
print(list1)
print(list1[-1])  # Accessing the last element
print(len(list1))  # Length of the list
a = len(list1)
print(a)


# Changing an element
print()
list1[1] = 100
print(list1)

# Adding elements
print()
list1.append("People")  # Adds to the end
list1.insert(2, "56")  # Inserts at index 2
print(list1)

# Removing elements
print()
list1.remove("Hello")  # Removes first occurrence of "Hello"
list1.pop()  # Removes the last element
list1.pop(2)  # Removes element at index 2
print(list1)
del list1[0]  # Deletes element at index 0
print(list1)
list1.clear()  # Clears the list
print(list1)
```

### Tuples

```python
# Tuples (Array in other languages)
# A tuple is a collection which is ordered and unchangeable. Allows duplicate members.

tuple1 = (23, 46, 78, "Hello", "World")
tup2 = (420, 9211 , "Sample", "Tuple")

print(type(tuple1))
print(tuple1[3])  # Accessing an element by index
print(tuple1)

print()
print(tuple1[-1])  # Accessing the last element
print(tuple1[1:4])  # Slicing the tuple
print(len(tuple1))  # Length of the tuple

# Tuples are immutable, so we cannot change, add, or remove elements
# tuple1[1] = 100  # This would raise an error
# tuple1.append("People")  # This would raise an error
# tuple1.remove("Hello")  # This would raise an error

tuple3 = tuple1 + tup2  # Concatenation of tuples
print(tuple3)  # Concatenation of tuples

```

### Set


```python
# Sets
# A set is a collection which is unordered and unindexed. No duplicate members.
set1 = {23, 46, 78, "abc", 4.6}

print(type(set1))
print(set1)  # Note: The order may vary
print(len(set1))  # Length of the set

print()
set1.add("new")  # Adding an element
print(set1)
set1.update([100, 200, "xyz"])  # Adding multiple elements
print(set1)

print()
set1.remove(46)  # Removing an element
print(set1)
set1.discard("abc")  # Removing an element
print(set1)

print()
a = set1.pop()  # Removes an arbitrary element
print(set1)
print(a)

# set1.clear()  # Clears the set

```

```python
# Set operations
seta = {1, 2, 3, 4, 5}
setb = {4, 5, 6, 7, 8}


print(seta.union(setb))  # Union (all unique elements)
print(seta.intersection(setb))  # Intersection (common elements)
print(seta.difference(setb))  # Difference (elements in seta not in setb)
print(setb.difference(seta))  # Difference (elements in setb not in seta)

print() #-- Adding a blank line for better readability
print(seta.symmetric_difference(setb))  # Symmetric Difference (elements in either seta or setb but not both)
print(seta.isdisjoint(setb))  # Check if sets have no elements in common
print(seta.issubset(setb))  # Check if seta is a subset of setb
print(seta.issuperset(setb))  # Check if seta is a superset of setb 


```
